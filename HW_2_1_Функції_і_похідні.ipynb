1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = sp.symbols('x')

f = 10*x - 5

f_prime = sp.diff(f, x)
print("Похідна функції:", f_prime)

x_vals = np.linspace(0, 10, 100)
f_vals = 10 * x_vals - 5
f_prime_vals = np.full_like(x_vals, 10) 

plt.figure(figsize=(8,5))
plt.plot(x_vals, f_vals, label='f(x) = 10x - 5')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 10", linestyle='--')
plt.xlabel("Години навчання (x)")
plt.ylabel("Оцінка")
plt.title("Графік функції та її похідної")
plt.legend()
plt.grid()
plt.show()
```

- Похідна функції — 10, тобто за кожну додаткову годину навчання оцінка збільшується на 10 балів.

- Лінія похідної є постійною горизонтальною прямою на рівні 10, тобто вплив кожної години на оцінку однаковий.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
f_sin = sp.sin(x)
f_sin_prime = sp.diff(f_sin, x)
print("Похідна sin(x):", f_sin_prime)

f_ln = sp.ln(x)
f_ln_prime = sp.diff(f_ln, x)
print("Похідна ln(x):", f_ln_prime)

# Графіки для sin(x)
x_vals_sin = np.linspace(0, 2*np.pi, 200)
f_vals_sin = np.sin(x_vals_sin)
f_prime_vals_sin = np.cos(x_vals_sin)

plt.figure(figsize=(8,5))
plt.plot(x_vals_sin, f_vals_sin, label='sin(x)')
plt.plot(x_vals_sin, f_prime_vals_sin, label="cos(x)", linestyle='--')
plt.title("Графік sin(x) та його похідної")
plt.legend()
plt.grid()
plt.show()

# Графіки для ln(x)
x_vals_ln = np.linspace(0.1, 10, 200)
f_vals_ln = np.log(x_vals_ln)
f_prime_vals_ln = 1 / x_vals_ln

plt.figure(figsize=(8,5))
plt.plot(x_vals_ln, f_vals_ln, label='ln(x)')
plt.plot(x_vals_ln, f_prime_vals_ln, label="1/x", linestyle='--')
plt.title("Графік ln(x) та його похідної")
plt.legend()
plt.grid()
plt.show()
```

- Для sin(x) похідна - це cos(x), графік похідної зміщений по фазі відносно синуса.

- Для ln(x) похідна — це 1/x, тому приріст зменшується при зростанні x.


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f_cost = x**2 - 5*x + 1
f_cost_prime = sp.diff(f_cost, x)
print("Похідна витрат:", f_cost_prime)

x_vals_cost = np.linspace(0, 100, 300)
f_vals_cost = x_vals_cost**2 - 5*x_vals_cost + 1
f_prime_vals_cost = 2*x_vals_cost - 5

plt.figure(figsize=(8,5))
plt.plot(x_vals_cost, f_vals_cost, label='Витрати: x² - 5x + 1')
plt.plot(x_vals_cost, f_prime_vals_cost, label="Похідна: 2x - 5", linestyle='--')
plt.xlabel("Кількість одиниць")
plt.ylabel("Витрати")
plt.title("Графік витрат і їхньої похідної")
plt.legend()
plt.grid()
plt.show()
```

- Похідна 2x - 5 показує, як змінюються витрати при зміні кількості вироблених одиниць.

- Функція витрат зростає квадратично, тобто чим більше виробляємо, тим сильніше ростуть витрати.
